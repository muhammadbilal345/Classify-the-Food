```python
df1 = df.drop(df[['Date', 'Evaporation', 'Sunshine', 'Cloud9am', 'Cloud3pm']], axis='columns')
```

```python
df1.head()
```

```python
df1.shape
```

```python
df2 = df1.dropna(subset=['MinTemp', 'MaxTemp', 'WindGustDir', 'WindDir9am', 'WindDir3pm', 'RainToday', 'RainTomorrow', 'Humidity9am', 'Humidity3pm', 'Temp3pm'])
```

```python
df2
```

```python
df2.Pressure9am.fillna(df2.Pressure9am.mean(), inplace=True)
df2.Pressure3pm.fillna(df2.Pressure3pm.mean(), inplace=True)
```

```python
missing = pd.DataFrame(df2.isnull().sum(), columns=['No. of missing values'])
missing['%missing_values'] = (missing/len(df2)).round(4)*100
missing
```

```python
df2.shape
```

```python
df2['RainToday'] = df2.RainToday.apply(lambda x: 1 if x == 'Yes' else 0)
df2['RainTomorrow'] = df2.RainTomorrow.apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in ['Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm']:
    df2[i] = le.fit_transform(df2[i])
df2.sample(5)
```

```python
df2.shape
```

```python
from sklearn.model_selection import train_test_split
X = df2.drop('RainTomorrow', axis='columns')
y = df2['RainTomorrow']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
X_train_scaled = scalar.fit_transform(X_train)
X_test_scaled = scalar.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)
```

```python
y_pred[:5]
```

```python
y_test[:5]
```

```python
model.score(X_test_scaled, y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
